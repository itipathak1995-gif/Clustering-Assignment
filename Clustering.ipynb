**Theoretical Questions**

1. What is unsupervised learning in the context of machine learning?

-Unsupervised learning is a type of machine learning where the model is trained on data without labeled outputs. The algorithm does not know the correct answers in advance. Instead, it analyzes the input data to find hidden patterns, structures, or relationships.

In unsupervised learning, there is no target variable. The model groups similar data points or reduces the number of features based on similarities or distance measures.

The two main types of unsupervised learning are:

Clustering – grouping similar data points together

Dimensionality Reduction – reducing features while keeping important information

It is commonly used in customer segmentation, anomaly detection, and pattern recognition.

2. How does K-Means clustering algorithm work?

- K-Means is a partition-based clustering algorithm that groups data into K clusters based on similarity.

It works in the following steps:

Choose K – Decide the number of clusters.

Initialize Centroids – Randomly select K data points as initial centroids.

Assign Points – Assign each data point to the nearest centroid using distance measures like Euclidean distance.

Update Centroids – Recalculate the centroid of each cluster by taking the mean of all assigned points.

Repeat – Repeat the assignment and update steps until centroids stop changing or convergence is achieved.

The goal of K-Means is to minimize the distance between data points and their respective cluster centroids.

3.Explain the concept of a dendrogram in hierarchical clustering ?

- A dendrogram is a tree-like diagram used to represent the results of hierarchical clustering. It shows how data points are merged or split into clusters at different levels of similarity.

In hierarchical clustering, clusters are formed step by step by combining the closest data points or clusters. The dendrogram visually displays this merging process.

The vertical axis represents the distance (or similarity) between clusters.

The horizontal axis represents the data points.

The height at which two clusters are joined indicates how similar or different they are.

By cutting the dendrogram at a certain height, we can decide the number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?
- The main difference between K-Means and Hierarchical Clustering lies in how clusters are formed.

K-Means:

Requires the number of clusters (K) to be specified in advance.

Uses centroids to form clusters.

Iteratively assigns points and updates centroids.

Suitable for large datasets.

Hierarchical Clustering:

Does not require the number of clusters initially.

Builds clusters step by step using a tree-like structure (dendrogram).

Can be agglomerative (bottom-up) or divisive (top-down).

More computationally expensive for large datasets.

5. What are the advantages of DBSCAN over K-Means?
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means:

No need to specify number of clusters in advance.

Can identify clusters of arbitrary shapes, while K-Means mainly works well with spherical clusters.

Can detect and handle noise and outliers effectively.

Works well when clusters have different sizes and densities.

In contrast, K-Means is sensitive to outliers and requires the value of K before clustering.

6. When would you use Silhouette Score in clustering?
- Silhouette Score is used to evaluate the quality of clustering. It measures how well each data point fits within its assigned cluster compared to other clusters.

It is mainly used:

To check how well clusters are separated.

To determine the optimal number of clusters (K).

To compare different clustering algorithms.

The score ranges from -1 to 1.
A value close to 1 indicates well-separated clusters, while values near 0 or negative suggest overlapping or poorly formed clusters.

7.  What are the limitations of Hierarchical Clustering?
- Hierarchical Clustering has several limitations:

High computational cost, especially for large datasets.

Once clusters are merged or split, the process cannot be reversed.

Sensitive to noise and outliers.

Choosing the correct number of clusters from the dendrogram can be difficult.

Performance depends on the choice of distance metric and linkage method.

8. Why is feature scaling important in clustering algorithms like K-Means?
- Feature scaling is important in K-Means because the algorithm uses distance measures (such as Euclidean distance) to form clusters.

If features are on different scales, variables with larger values will dominate the distance calculation. This can lead to incorrect cluster formation.

By applying feature scaling methods like Standardization or Normalization, all features contribute equally to the clustering process, resulting in more accurate and meaningful clusters.

9.  How does DBSCAN identify noise points?
- DBSCAN identifies noise points based on density criteria using two parameters:

Epsilon (ε) – maximum distance between two points

MinPts – minimum number of points required to form a dense region

A point is considered a core point if it has at least MinPts neighbors within ε distance.
If a point is not a core point and does not belong to the neighborhood of any core point, it is labeled as a noise (outlier) point.

Thus, points that do not belong to any dense cluster are automatically classified as noise.

10. Define inertia in the context of K-Means?
- Inertia in K-Means refers to the sum of squared distances between each data point and its assigned cluster centroid.

It measures how tightly the data points are grouped within a cluster. Lower inertia indicates more compact and well-formed clusters.

K-Means aims to minimize inertia during the clustering process.

11. What is the elbow method in K-Means clustering?
- The Elbow Method is used to determine the optimal number of clusters (K) in K-Means clustering.

It works by running K-Means for different values of K and calculating the inertia (sum of squared distances) for each value.

A graph is plotted between the number of clusters (K) and inertia. As K increases, inertia decreases. The point where the decrease in inertia becomes slow and forms an "elbow" shape is considered the optimal number of clusters.

This point represents a balance between cluster compactness and simplicity.

12. Describe the concept of "density" in DBSCAN?
- In DBSCAN, density refers to the number of data points within a specified radius (ε) around a given point.

A region is considered dense if it contains at least a minimum number of points (MinPts) within the ε distance.

Based on density, DBSCAN classifies points into:

Core points – have at least MinPts neighbors within ε.

Border points – have fewer than MinPts neighbors but are within the neighborhood of a core point.

Noise points – do not belong to any dense region.

Clusters are formed by connecting dense regions together.

13.  Can hierarchical clustering be used on categorical data?
- Yes, hierarchical clustering can be used on categorical data, but it requires an appropriate similarity or distance measure.

Since standard distance measures like Euclidean distance are suitable for numerical data, categorical data needs measures such as Hamming distance or other similarity-based metrics.

Hierarchical clustering is flexible because it allows the use of different distance metrics and linkage methods, making it suitable for both numerical and categorical datasets.

14. What does a negative Silhouette Score indicate?
- A negative Silhouette Score indicates that a data point may have been assigned to the wrong cluster.

It means the point is closer to data points in a neighboring cluster than to those in its own cluster. This suggests poor clustering or overlapping clusters.

A negative value usually reflects incorrect cluster separation or an inappropriate number of clusters.

15. Explain the term "linkage criteria" in hierarchical clustering?
- Linkage criteria refer to the method used to measure the distance between clusters in hierarchical clustering.

It determines how the distance between two clusters is calculated when merging them.

Common types of linkage criteria include:

Single Linkage – distance between the closest points of two clusters.

Complete Linkage – distance between the farthest points of two clusters.

Average Linkage – average distance between all pairs of points in two clusters.

Ward’s Method – minimizes the total within-cluster variance.

The choice of linkage criteria affects the shape and structure of the final clusters.

16.  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
- K-Means may perform poorly on data with varying cluster sizes or densities because it assumes that clusters are spherical and of similar size.

Since K-Means uses the mean (centroid) and Euclidean distance, it tries to create clusters with equal variance. When clusters have different sizes or densities, the algorithm may incorrectly split dense clusters or merge sparse ones.

It is also sensitive to outliers, which can shift the centroid and affect cluster formation.

17. What are the core parameters in DBSCAN, and how do they influence clustering?
- The two core parameters in DBSCAN are:

Epsilon (ε) – Defines the maximum distance between two points to be considered neighbors.

MinPts – Minimum number of points required within ε to form a dense region.

These parameters influence clustering as follows:

A small ε may result in many small clusters or more noise points.

A large ε may merge distinct clusters into one.

A higher MinPts requires denser regions to form clusters.

A lower MinPts may create clusters from sparse regions.

Choosing appropriate values for ε and MinPts is important for meaningful clustering results.

18. How does K-Means++ improve upon standard K-Means initialization?
- K-Means++ improves the initialization of centroids in standard K-Means to achieve better clustering results.

Instead of selecting initial centroids randomly, K-Means++ selects the first centroid randomly and then chooses the remaining centroids based on probability, giving higher chance to points that are far from existing centroids.

This ensures that initial centroids are well separated, which:

Reduces the chances of poor clustering

Improves convergence speed

Often leads to lower inertia

Thus, K-Means++ provides more stable and accurate clustering compared to standard random initialization.

19. What is agglomerative clustering?
- Agglomerative clustering is a type of hierarchical clustering that follows a bottom-up approach.

In this method, each data point starts as its own individual cluster. The algorithm then repeatedly merges the two closest clusters based on a chosen distance measure and linkage criteria.

This process continues until all data points are merged into a single cluster or until the desired number of clusters is achieved.

The result of agglomerative clustering is usually represented using a dendrogram.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
- Silhouette Score is considered better than inertia because it evaluates both cluster cohesion and separation.

Inertia only measures how close data points are to their own cluster centroid. It always decreases as the number of clusters increases, which makes it difficult to determine the optimal number of clusters.

Silhouette Score, on the other hand:

Measures how well a point fits within its cluster

Compares it with neighboring clusters

Ranges from -1 to 1, making interpretation easier

Thus, Silhouette Score provides a more balanced evaluation of clustering quality than inertia alone.

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart?

```python
29.
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Store silhouette scores
k_values = range(2, 6)
scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot bar chart
plt.figure()
plt.bar(k_values, scores)

plt.title("Silhouette Scores for Different K Values")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries?

```python
31.
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=300,
                  centers=3,
                  cluster_std=2.5,   # High std → overlapping clusters
                  random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Create mesh grid for decision boundary
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster for each grid point
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            marker='X',
            s=200)

plt.title("K-Means with Decision Boundaries (Overlapping Clusters)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python
44.
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load Iris dataset
iris = load_iris()
X = iris.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans (3 clusters for Iris)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Compute silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X_scaled, labels)

# Plot silhouette values
plt.figure()
y_lower = 10

for i in range(3):
    ith_cluster_values = sample_silhouette_values[labels == i]
    ith_cluster_values.sort()

    size_cluster_i = ith_cluster_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0,
                      ith_cluster_values)

    y_lower = y_upper + 10

plt.title("Silhouette Plot for Iris Dataset (KMeans)")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.axvline(x=silhouette_score(X_scaled, labels))
plt.show()
```

47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count

```python
47.
# Import required libraries
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Generate noisy blob data
X, y = make_blobs(n_samples=400,
                  centers=3,
                  cluster_std=2.0,   # higher std → more noise/overlap
                  random_state=42)

# Scale data (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (excluding noise label -1)
unique_labels = set(labels)
num_clusters = len(unique_labels - {-1})

# Count noise points
noise_count = np.sum(labels == -1)

print("Number of clusters (excluding noise):", num_clusters)
print("Number of noise points identified:", noise_count)
```

48.  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
